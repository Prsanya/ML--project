Problem Statement:
The dataset is in Zipped format, we have to extract the dataset in the program, visualize the number of
male and female babies born in a particular year, and find out popular baby names.

Description: This project not only focusses on implementing data manipulation and data visualization
using Pandas library, but also tests your ability to deal with real word problem statements.

Dataset: Popular baby names data provided by Social Security Administration (SSA) of United States 
   

```python
import pandas as pd
years=range(1880,2021)
li=[]
colums=["name","sex","birth"]

for year in years:
    df=pd.read_csv("C:/Users/PRSANYA/Desktop/Names/yob%d.txt"%year,names=colums)
    df["year"]=year
    li.append(df)
name=pd.concat(li,axis=0)    
name
```

```python
f=name.groupby(["year","sex"], group_keys=False)[["birth"]].sum().reset_index().sort_values(by="birth",ascending=False)

top=f.head(100)
top
```

```python
#Visualize the number of male and female babies born in a particular year 
p=f.pivot_table(index="year",columns="sex",values="birth", aggfunc = sum).fillna(0)
p
```

```python
p.plot()
```

```python
#Analyse baby names by sorting out top 100 birth counts 
s=name.sort_values(by="birth",ascending=False)
s
```

```python
#Group them by names to find out popular baby names 
t_100=s.head(100)
```
